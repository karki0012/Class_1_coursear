# Gradient Descent for Linerar Regression
## Goals
- In this we will automate the process of optimizing *w* and *b* using gradient descent

```python
import numpy as np 
import pandas as pd 

# importing data 
df = pd.read_excel (r'D:\\article_writting\\efficiency\\Data_result\\Local_Level_Data.xlsx')

# Data assign
x_train = df['N_of_candt'] # features 
y_train = df["Cr_expn"] # target values
```

```python
# function to calculate the cost
def compute_cost ( x, y, w, b):
    m = x.shape[0]
    cost = 0

    for i in range (m):

       f_wb = w * x[i] + b
       cost = cost + (f_wb - y[i])**2
    
       total_cost = 1/(2 * m) * cost
     
       return total_cost
```

## Implement Gradient Descent
- compute_gradient 
- compute_cost
- gradiemt_descent, utilizing compute_gradient and compute_cost

```python
# Compute gradient 
def compute_gradient (x, y, w, b):

    # Number of training examples
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0
    
    for i in range (m):
        f_wb = w*x[i] + b
        dj_dw_i = (f_wb - y[i])*x[i]
        dj_db_i = f_wb - y[i]
        dj_db += dj_db_i
        dj_dw += dj_dw_i

        dj_dw = dj_dw/m
        dj_db = dj_db/m

        return dj_dw, dj_db

```

```python
 # Gradient Descent computation

def gradient_descent (x, y, w_init, b_init, alpha, num_iters):
    
    w = w_init
    b = b_init
    for i in range (num_iters):
        dj_dw, dj_db = compute_gradient (x, y, w, b)

        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        
        return w, b 
    
```

```python
# Initialize parameters 
w_init = 0
b_init = 0
# some gradient descent settings
alpha = 1.0e-6
iterations = 10000

# Run gradient descent 
w_final, b_final = gradient_descent(x_train, y_train, w_init, b_init, alpha, iterations)

print(f"(w,b) found by gradient descent: ({w_final:8.4f},{b_final:8.4f})")
```
